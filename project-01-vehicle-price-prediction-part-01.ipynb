## Features Enginering(Round 01)

```python
df.info()
```

### Clearnig features data ( CC and kmpl from releven columns)

```python
df.head(3)
```

```python
df["mileage"]= df["mileage"].str.replace(" kmpl","0")
df["mileage"]= df["mileage"].str.replace(" km/kg","0")

```

```python
df.head(3)
```

```python
df["engine"]= df["engine"].str.replace("[ CC]","")
```

```python
df.head(3)
```

```python
df["max_power"]= df["max_power"].str.replace("[ bhp]","")
```

```python
df.head(3)
```

### Convert worng Object data types into correct data types

```python
df['mileage'] = df["mileage"].astype(float)
```

```python
df['engine'] = df["engine"].astype(int)
```

```python
df['max_power'] = df["max_power"].astype(float)
```

```python
df.info()
```

 ## Feature Selection ( Round 01 )

```python
# target variable 
y = df["selling_price"]
```

```python
# Filtering unwanted features
df.drop("name", axis=1 , inplace= True)
df.drop("selling_price", axis=1, inplace=True)
df.drop("torque", axis=1, inplace=True)
```
